```python
# lab 2.1
#构建 最上层train 第二层cost 第三层hypothesis 第四层 w和b 
#运行train相当于对w和b进行最优化
#首先建立graph
import tensorflow as tf 
x_train=[1,2,3]
y_train=[1,2,3]

w=tf.Variable(tf.random_normal([1]),name='weight') #Variable 定义未知数和名称
b=tf.Variable(tf.random_normal([1]),name='bias')
hypothesis=x_train*w+b #graph建立到这

cost=tf.reduce_mean(tf.square(hypothesis-y_train)) # tf.reduce_mean 求平均值
train=tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost) #GradientDescentOptimizer 要大写
# optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.01)
# train=optimizer.minize(cost) 与上句一样

sess=tf.Session() #将建立的graph载入到一个会话session里面
sess.run(tf.global_variables_initializer()) #构建完整个模型并在会话中加载模型后，初始化所有变量
for step in range(2001):
 sess.run(train)
 if step % 20 == 0:
        print(step,sess.run(cost),sess.run(w),sess.run(b)) #加上sess.run才可以把session里面的某个数值显示，否则返回结构信息
```

```python
#lab 2.2
w=tf.Variable(tf.random_normal([1]),name='weight')
b=tf.Variable(tf.random_normal([1]),name='bias')
X=tf.placeholder(tf.float32,shape=[None])
Y=tf.placeholder(tf.float32,shape=[None]) #None 代表可以放一个数值或多个数值

hypothesis=X*w+b
cost=tf.reduce_mean(tf.square(hypothesis - Y))
train=tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

#Launch the graph in a session
sess=tf.Session()
sess.run(tf.global_variables_initializer())

#与上面的例子不同，这里直接一次性运行session整个模型
for step in range(2001):
    cost_val,w_val,b_val,_=sess.run([cost,w,b,train],
              feed_dict={X:[1,2,3,4,5], 
               Y:[2.1,3.1,4.1,5.1,6.1]})
if step % 20 == 0: # if 顶格的话只能缩进，输出最后的结果  不顶格输出全部结果
        print(step,cost_val,w_val,b_val)
        
#testing the model
print(sess.run(hypothesis,feed_dict={X:[5]}))
print(sess.run(hypothesis,feed_dict={X:[6,7]}))
```
