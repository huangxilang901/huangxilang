## （2）数据操作

（1）读取香港酒店数据。

```python
hk = pd.read_excel(r'C:\Users\Administrator\Desktop\作业3\香港酒店数据.xlsx')
hk[:3]
```

（2）按照数据的内容，重新设置数据的索引，重新设置列名称为'名字','类型','城市','地区','地点','评分','评分人数','价格'。

```python
hk.columns = ['名字','类型','城市','地区','地点','评分','评分人数','价格']
hk.index.name='索引'
hk[:3]
```

（3）查看所有类型为“浪漫情侣”的酒店

```python
hk[hk['类型']=='浪漫情侣'][:3]
```

（4）查看所有类型为“浪漫情侣”，地区在湾仔的酒店

```python
hk[(hk.类型=='浪漫情侣')&(hk.地区=='湾仔')][:10]
```

（5）查看所有地址在观塘或者油尖旺，评分大于4的酒店

```python
hk[((hk.地区=='油尖旺')|(hk.地区=='观塘'))&(hk.评分>4)][:10]
```

（6）查看类型缺失的数据

```python
hk[hk['类型'].isnull()]
```

（7）用“其他”填充类型和地区

```python
hk['类型'].fillna('其他',inplace=True)
hk[hk['类型'].isnull()]
```

（8）用评分均值填充缺失值

```python
hk['评分'].fillna(np.mean(hk['评分']),inplace=True)
```

（9）删除价格和评分人数的缺失值

```python
hk.dropna(subset=['价格', '评分人数'],inplace=True)
```

（10）保存到“酒店数据1.xlsx”

```python
hk.to_excel('hotel_data_new2.xlsx')
```
